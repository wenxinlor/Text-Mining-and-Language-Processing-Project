```python
#Load Data
import pandas as pd

df_all = pd.read_csv('Suicide_Detection.csv')
```

```python
df = df_all[0:100000]
print(len(df))
```

```python
df_suicide = list(df.loc[df["class"] == "suicide", "text"])
df_non_suicide = list(df.loc[df["class"] == "non-suicide", "text"])
joined = list(df["text"])
```

```python
joined_bigram_before = [(x + "_" + str(i).split()[j + 1]) for i in joined for j, x in enumerate(str(i).split()) if j < len(str(i).split()) - 1]
print("2")
joined_trigram_before = [(x + "_" + str(i).split()[j + 1] + "_" + str(i).split()[j+2]) for i in joined for j, x in enumerate(str(i).split()) if j < len(str(i).split()) - 2]
print("3")
```

```python
#list for bigrams + trigrams
joined_bi_tri_grams = []
#no code as words will be added to the list from the 2 codes below
```

```python
#list for bigrams
joined_bigram = []
for bigram in joined_bigram_before:
    split_word = bigram.split("_")
    x = split_word[0]
    y = split_word[1]
    if len(x) > 2 and len(y) > 2:
        word = x + "_" + y
        joined_bigram.append(word) 
        joined_bi_tri_grams.append(word)
```

```python
#list for trigrams
joined_trigram = []
for trigram in joined_trigram_before:
    split_word = trigram.split("_")
    x = split_word[0]
    y = split_word[1]
    z = split_word[2]
    if len(x) > 2 and len(y) > 2 and len(z) > 2:
        word = x + "_" + y + "_" + z
        joined_trigram.append(word)
        joined_bi_tri_grams.append(word)
```

```python
#imports for LDA and coherence model
!pip install gensim
import nltk
from gensim.utils import simple_preprocess
import gensim
from gensim.models import CoherenceModel
```

```python
import gensim.corpora as corpora

#create bigram corpus
joined_words_bigram = [str(d_s).split() for d_s in joined_bigram]
print(joined_words_bigram)

# Create Dictionary
id2word_joined_bigram = corpora.Dictionary(joined_words_bigram)
# Create Corpus
texts_joined_bigram = joined_words_bigram
# Term Document Frequency
joined_corpus_bigram = [id2word_joined_bigram.doc2bow(text1) for text1 in texts_joined_bigram]


#create trigram corpus
joined_words_trigram = [str(d_s).split() for d_s in joined_trigram]
id2word_joined_trigram = corpora.Dictionary(joined_words_trigram)
texts_joined_trigram = joined_words_trigram
joined_corpus_trigram = [id2word_joined_trigram.doc2bow(text1) for text1 in texts_joined_trigram]


#create bigram + trigram corpus
joined_words_both = [str(d_s).split() for d_s in joined_bi_tri_grams]
id2word_joined_both = corpora.Dictionary(joined_words_both)
texts_joined_both = joined_words_both
joined_corpus_both = [id2word_joined_both.doc2bow(text1) for text1 in texts_joined_both]
```

```python
#bigram lda model
from pprint import pprint

# number of topics
num_topics = 10
# Build LDA model
lda_model_joined_bigram = gensim.models.LdaMulticore(corpus=joined_corpus_bigram, id2word=id2word_joined_bigram, num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model_joined_bigram.print_topics())
doc_lda_joined_bigram = lda_model_joined_bigram[joined_corpus_bigram]
```

```python
#trigram lda model
from pprint import pprint

# number of topics
num_topics = 10
# Build LDA model
lda_model_joined_trigram = gensim.models.LdaMulticore(corpus=joined_corpus_trigram, id2word=id2word_joined_trigram, num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model_joined_trigram.print_topics())
doc_lda_joined_trigram = lda_model_joined_trigram[joined_corpus_trigram]
```

```python
#bigram + trigram lda model
from pprint import pprint

# number of topics
num_topics = 10
# Build LDA model
lda_model_joined_both = gensim.models.LdaMulticore(corpus=joined_corpus_both, id2word=id2word_joined_both, num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model_joined_both.print_topics())
doc_lda_joined_both = lda_model_joined_both[joined_corpus_both]
```

```python
#saving the models
doc_lda_joined_bigram.save('lda.model_joined_bigram')
doc_lda_joined_trigram.save('lda.model_joined_trigram')
doc_lda_joined_both.save('lda.model_joined_bigram_trigram')
```

```python
#cohernece score for bigram
coherence_model_lda_joined_bigram = CoherenceModel(model=lda_model_joined_bigram, texts=texts_joined_bigram, dictionary=id2word_joined_bigram, coherence='c_v')

coherence_lda_joined_bigram = coherence_model_lda_joined_bigram.get_coherence()

print('\nCoherence Score Bigram: ', coherence_lda_joined_bigram)
```

```python
#cohernece score for trigram
coherence_model_lda_joined_trigram = CoherenceModel(model=lda_model_joined_trigram, texts=texts_joined_trigram, dictionary=id2word_joined_trigram, coherence='c_v')

coherence_lda_joined_trigram = coherence_model_lda_joined_trigram.get_coherence()

print('\nCoherence Score Trigram: ', coherence_lda_joined_trigram)
```

```python
#cohernece score for bigram + trigram
coherence_model_lda_joined_both = CoherenceModel(model=lda_model_joined_both, texts=texts_joined_both, dictionary=id2word_joined_both, coherence='c_v')

coherence_lda_joined_both = coherence_model_lda_joined_both.get_coherence()

print('\nCoherence Score bigram + trigram: ', coherence_lda_joined_both)
```
